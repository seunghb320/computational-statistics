```python
import random
random.seed(1)
from scipy.stats import norm
```

```python
def roll():
  return random.randint(1,6)
def simulation():
  count = []
  for i in range(100):
    dice = roll()
    count.append(dice)
  return np.mean(count)
def repeat():
  xbars = []
  for i in range(10000):
    xbar = simulation()
    xbars.append( xbar)
  mean = np.mean(xbars)
  var = np.var(xbars)
  sd = np.std(xbars)
  print('평균:',mean)
  print('분산:',var)
  pdf_plot = np.linspace(2.5,4.5,1000)
  plt.figure(figsize=(10,6))
  plt.hist(xbars, density=True)
  plt.plot(pdf_plot, norm.pdf(pdf_plot, loc=mean, scale=sd))
```

```python
repeat()
```
